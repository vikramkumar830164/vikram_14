```python
# Import the star of our show
import seaborn as sns
# We still need Matplotlib for customizations
import matplotlib.pyplot as plt
import pandas as pd

# Load the data - hamara wahi purana dost, Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Basic data cleaning
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.dropna(subset=['Embarked'], inplace=True)

print("Seaborn is ready! Let the magic begin.")
```

```python
# Let's try the 'darkgrid' style
sns.set_theme(style="darkgrid")

# Now, let's make a simple histogram. Notice how it looks better than the default Matplotlib one.
plt.hist(df['Age'], bins=20)
plt.title("A Matplotlib Plot with Seaborn's Magic Touch")
plt.show()

# Let's reset to the default theme for our lecture
sns.set_theme(style="whitegrid") # A clean, popular choice
```

```python
# Seaborn automatically makes your plots look better, but you can also set a theme for all your plots with one line of code.
```

```python
# The biggest advantage of Seaborn is that its functions are designed to work directly with Pandas DataFrames.
# You just tell it which columns to use for x and y.
```

```python
# Create a histogram of Age using Seaborn
sns.histplot(data=df, x='Age', bins=30, kde=True) # kde = Kernel Density Estimate

plt.title('Age Distribution with Seaborn')
plt.show()
```

```python
# Create a box plot to compare Age across different Classes
plt.figure(figsize=(10, 6)) # You can still use plt to set figure size
sns.boxplot(data=df, x='Pclass', y='Age')

plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
# Look at that! No need to manually separate data for each class. Seaborn does it for you.  SMART WORK
```

```python
# Count the number of passengers from each port
sns.countplot(data=df, x='Embarked', palette='viridis') # palette changes the color scheme

plt.title('Passenger Count by Port of Embarkation')
plt.show()
```

```python
# countplot - Bar Chart for Counts
# This is the direct replacement for value_counts().plot(kind='bar').
```

```python
#  The Statistical Bar Chart
# Important: Seaborn's barplot is different. It automatically groups your data by the categorical variable and plots the mean of the numerical variable.
```

```python
# Seaborn will automatically calculate the average fare for males and females
sns.barplot(data=df, x='Sex', y='Fare')

plt.title('Average Fare Paid by Gender')
plt.show()
```

```python
# scatterplot with hue
# # The hue parameter is Seaborn's secret weapon. It lets you add a third variable to your plot, represented by color.
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')

plt.title('Age vs. Fare, colored by Survival Status')
plt.show()
```

```python
# heatmap - The Correlation Detective
# A heatmap is perfect for visualizing a correlation matrix. It shows how strongly variables are related to each other.
```

```python
# First, we need to calculate the correlation matrix for numerical columns
correlation_matrix = df.corr(numeric_only=True)

# Now, create the heatmap
plt.figure(figsize=(12, 9))
sns.heatmap(
    correlation_matrix,
    annot=True,       # Show the correlation values on the map
    cmap='viridis',     # Choose a color map
    fmt='.2f'         # Format numbers to 2 decimal places
)

plt.title('Correlation Matrix of Titanic Features')
plt.show()
```

```python
# pairplot - The Ultimate EDA Shortcut
# What if you want to see the relationship between all numerical variables at once? The pairplot does exactly that.
```

```python
#---------------------------------------- EDA   KA   BRAHMASTRA ------------------------------------------------------------
```

```python
# Create a pairplot for a subset of columns to keep it clean
# We'll also use 'hue' to see how survival differs
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='husl')

plt.show()
```

```python
# What you are seeing:
# The plots on the diagonal are histograms (or KDEs), showing the distribution of each single variable.
# The plots off the diagonal are scatter plots, showing the relationship between every pair of variables.
# The hue parameter colors everything by survival status.
# In just one line of code, we have performed a huge chunk of our EDA!
```
